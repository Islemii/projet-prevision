## ARDL

```r
install.packages("ARDL")
library(ARDL)
```

```r
# Set the data as a time series, excluding the `t` column
my_data3_ts <- ts(my_data3[,-1], start = 2005, end = 2014)
```

```r
print(my_data3_ts)
```

### Nombre max de retard lag

```r
# Fit ARDL model with yt as the dependent variable
model <- ardl(yt ~ xt + ht, data = my_data3_ts, order = c(1,1,1))  # Order specifies lags for each variable

# Print the model summary
summary(model)
```

```r
residuals <- residuals(model)  # Obtenir les résidus du modèle
```

```r
jarque.bera.test(residuals)
```

```r
# Effectuer le test de Durbin-Watson
dwtest(model)
```

```r
shapiro.test(residuals)
```

```r
library(lmtest)
bgtest(model)
```

```r
library(lmtest)
bptest(model)
```

```r
# Modele ARMA
set.seed(123)  # Fixe la graine pour reproductibilité

# Génération d'une série ARMA(1,1)
library(forecast)
xt <- arima.sim(model = list(order = c(1, 0, 1), ar = 0.7, ma = 0.5), n = 100)

# Affichage de la série
plot(xt, type = "l", col = "blue", main = "Série Stationnaire (ARMA(1,1))", ylab = "x_t", xlab = "Temps")
```

```r
library(forecast)

# Ajustement du modèle ARMA(1,1)
fit <- Arima(xt, order = c(1, 0, 1))
summary(fit)
```

```r
# Prévoir les 10 prochaines valeurs
forecasted <- forecast(fit, h = 10)

# Affichage des prévisions
plot(forecasted, main = "Prévisions ARMA(1,1)", ylab = "x_t", xlab = "Temps")
```

```r
# Valeur prédite pour n = 110
forecasted$mean[10]
```

```r
# Step 4: Print the specific value for n = 110
cat("Predicted value for n = 110:", forecasted$mean[10], "\n")
```

```r
set.seed(123)
xt <- rnorm(100, mean = 0, sd = 1)  # Moyenne = 0, Écart-type = 1

# Affichage de la série
plot(xt, type = "l", col = "blue", main = "Série Stationnaire (Bruit Blanc)", ylab = "x_t", xlab = "Temps")
abline(h = mean(xt), col = "red", lty = 2)  # Ligne représentant la moyenne
```

```r
set.seed(123)
xt <- arima.sim(model = list(order = c(1, 0, 1), ar = 0.5, ma = 0.3), n = 100)

# Affichage de la série
plot(xt, type = "l", col = "purple", main = "Série Stationnaire (ARMA(1,1))", ylab = "x_t", xlab = "Temps")
```

```r
set.seed(123)
n <- 100
xt_non_stationary <- cumsum(rnorm(n))  # Somme cumulative d'un bruit blanc (non stationnaire)

# Affichage de la série
plot(xt_non_stationary, type = "l", col = "red", main = "Série Non Stationnaire", ylab = "x_t", xlab = "Temps")
```

```r
# Test ADF
adf.test(xt_non_stationary)
```

```r
library(forecast)

# Ajuster le modèle ARIMA
fit_non_stationary <- auto.arima(xt_non_stationary)

# Résumé du modèle
summary(fit_non_stationary)
```

```r
# Étape 3 : Prévoir les 10 prochaines valeurs
forecasted_ns <- forecast(fit_non_stationary, h = 10)

# Étape 4 : Graphique combiné
plot(forecasted_ns, 
     main = "ARIMA sur une Série Non Stationnaire", 
     ylab = "x_t", 
     xlab = "Temps", 
     col = "blue")

# Ajout de la série originale au graphique
lines(1:n, xt_non_stationary, col = "red", lwd = 2)  # Série originale
legend("topleft", legend = c("Prévisions", "Série Originale"), 
       col = c("blue", "red"), lty = 1, lwd = 2)
```

```r
fit2 <- Arima(xt_non_stationary, order = c(0, 2, 0))
summary(fit2)
```

```r
# Étape 3 : Prévoir les 10 prochaines valeurs
forecasted_ns2 <- forecast(fit2, h = 10)
# Étape 4 : Graphique combiné
plot(forecasted_ns2, 
     main = "ARIMA sur une Série Non Stationnaire", 
     ylab = "xt_non_stationary", 
     xlab = "Temps", 
     col = "blue")

# Ajout de la série originale au graphique
lines(1:n, xt_non_stationary, col = "red", lwd = 2)  # Série originale
legend("topleft", legend = c("Prévisions", "Série Originale"), 
       col = c("blue", "red"), lty = 1, lwd = 2)
```
